## 8.3 中国股票交易数据项目实战
### 8.3.1 任务分析
- 逐行读入单只股票
- 利用字典存储单只股票信息
- 利用嵌套字典存储多只股票信息
- 股票信息保存到文件

### 8.3.2 任务实现
#### 逐行读入单只股票


```python
with open("data/StockData.txt", "r") as f:
    lines = f.readlines()
print(lines[1])
```

#### 利用字典存储单只股票信息


```python
stock_list = []
for line in lines:
    stock = eval(line)
    
    cjl = stock["成交量"][:-2]
    if cjl.strip() == "":
        stock["成交量"] = 0
    else:
        stock["成交量"] = float(cjl) * 10000
        
    unit = stock["成交额"][-1]
    cje = stock["成交额"][:-1]
    if cje.strip() == "-":
        stock["成交额"] = 0
    else:
        if unit == "万":
            stock["成交额"] = float(cje) * 10000
        else:
            stock["成交额"] = float(cje) * 100000000
            
    wb = stock["委比"][:-1]
    if wb.strip() == "-":
        stock["委比"] = 0
    else:
        stock["委比"] = round(float(wb) / 100, 2)
        
    stock_list.append(stock)
    
print(stock_list[1])
```

#### 利用嵌套字典存储多只股票信息


```python
from pprint import pprint
stock_dict = {}
for dic in stock_list:
    stock_dict[dic["股票名称"]] = dic
pprint(stock_dict)
```

```python
print(stock_dict["万东医疗"])
```

#### 股票信息保存到文件

```python
import pickle
f = open("data/股票信息.pkl", "wb")
pickle.dump(stock_dict, f)
f.close()
```

```python
# json文件
import json
f = open("data/stcok.json", "w")
json.dump(stock_dict,f)
f.close()
```

### 8.5.1 列表转化字典

```python
from  datetime import datetime 
import locale
locale.setlocale(locale.LC_CTYPE,"chinese")

with open("data/宁波精达.txt", "r") as f:
    lines = f.readlines()
lines = [x[:-2] for x in lines]
head = lines[0].split(",")
lines = lines[1:]
#print(lines)
#print(head)
stocks = {}
for line in lines:
    stock = {}
    line_list = line.split(",")
    for i in range(3, len(head)):
        stock[head[i]] = line_list[i]
    stock_date = line_list[0].split(" ")[0]
    stock_date = datetime.strptime(stock_date, "%Y-%m-%d").strftime("%Y年%m月%d日")
    stocks[stock_date] = stock
    
pprint(stocks)
```
