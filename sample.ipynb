```python
import pandas as pd
df = pd.read_excel("data/招聘-数据分析.xlsx")
```

```python
df["薪水单位"] = df["薪水"].str.extract("^.*?(.{1})/.*?")
df["薪水单位"] = df["薪水单位"].str.replace("万","10").replace("千","1")
df.head()
```

```python
df["薪水-高"] = df["薪水"].str.extract("^(.*?)-.*?")
df.head()
```

```python
df["薪水-低"] = df["薪水"].str.extract("^.*?-(.*?)[万|千].*?")
df.head()
```
